```python
import seaborn as sns
%matplotlib inline 
tips = sns.load_dataset('tips')
flights = sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
fc = flights.corr()
```

```python
sns.heatmap(tc, annot=True,cmap='coolwarm')
```

```python
fc = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fc,annot=True,fmt="d",cmap='coolwarm',linewidths=1,linecolor='white')
```

```python
sns.clustermap(fc,standard_scale=1,cmap='coolwarm')
```
