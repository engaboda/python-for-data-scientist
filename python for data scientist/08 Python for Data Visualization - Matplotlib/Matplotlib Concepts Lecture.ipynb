```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#this will let you see your ploting in jupyter 
```

```python
import numpy as np
x =np.linspace(0,5,11)
y = x**2
```

```python
#--------------there are two way to make plot 
#first using Functional Method
#second using Object Oriented Method
```

#--------------------------------Functional--------------------------------

```python
plt.plot(x,y,'r')
#plt.plot(x,y,'r--')
#plt.plot(x,y,'r-')
```

```python
plt.plot(x,y,'r')
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Plotting A and Y ")
```

#Createing Multi Plot in the same Shape

```python
#num of rows , numof columns ,
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#---------------------------------OO------------------------------
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g')
axes.set_ylabel("Y Label")
axes.set_xlabel("X Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.8,0.8,0.2,0.2])
#add_axes(x0,y0,width,height)
#x0 -------------------
#y0  |
#   |
axes1.plot(x,y,'r')
axes1.set_title("Large Plot")
axes2.plot(y,x,'b')
axes2.set_title("Small Plot")
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#plt.tight_layout()#to Solve overlapping
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("Fisrt")
axes[1].plot(y,x)
axes[1].set_title("Second")
```

```python
fig = plt.figure(figsize=(3,2))
#figsize=(width,height) 
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(8,8))
axes[0][0].plot(x**3,y,label='X Cubed')
axes[0][0].plot(x**2,x, label='X Squered')
axes[0][1].plot(y,y)
axes[1][0].plot(x,y)
axes[1][1].plot(x,x)
plt.tight_layout()
axes[0][0].legend() # we should give each axes label to use legend
#if legend inbad location we can use 
#axes.legend(loc=0)
#‘best’	0
#‘upper right’	1
#‘upper left’	2
#‘lower left’	3
#‘lower right’	4
#‘right’	5
#‘center left’	6
#‘center right’	7
#‘lower center’	8
#‘upper center’	9
#‘center’	10
```

```python
#to save a figure
fig.savefig("my_fig.png",dpi=500)
```
