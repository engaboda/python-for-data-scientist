```python
import pandas as pd
```

```python
df = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3'],
                  } , index=[0,1,2,3])
```

```python
df1 = pd.DataFrame({'A':['A00','A10','A20','A30'],
                   'B':['B00','B10','B20','B30'],
                   'C':['C00','C10','C20','C30'],
                   'D':['D00','D10','D20','D30'],
                  } , index=[4,5,6,7])
```

```python
df1
```

```python
df2 = pd.DataFrame({'A':['A000','A100','A200','A300'],
                   'B':['B000','B100','B200','B300'],
                   'C':['C000','C100','C200','C300'],
                   'D':['D000','D100','D200','D300'],
                  } , index=[8,9,10,11])
```

```python
df2
```

```python
#Concatination basicully  glues together DataFrame Keep in mind That
#dimensions should match along the axis you are concatinating on You can use (pd.concat())
```

```python
pd.concat([df,df1,df2])
```

```python
pd.concat([df,df1,df2],axis=1)
```

```python
left = pd.DataFrame({'Key':['k0','k1','k2','k3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3'],
                    } )
```

```python
left
```

```python
right = pd.DataFrame({'Key':['k0','k1','k2','k3'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3'],
                    } )
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='Key')
```

```python
left = pd.DataFrame({'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3'],
                    },index=['k0','k1','k2','k3'] )

right = pd.DataFrame({'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3'],
                    },index=['k0','k1','k2','k3'] )
```

```python
left.join(right)
```
